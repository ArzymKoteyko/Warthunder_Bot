```python
import time
from multiprocessing import Process
import pycurl
from io import BytesIO
import json
import math
import time
from multiprocessing.pool import ThreadPool
import ctypes
import scipy.interpolate
import numpy as np
import win32api

SendInput = ctypes.windll.user32.SendInput

KeyCodes = {'W'     : 0x11,
            'A'     : 0x1E,
            'S'     : 0x1F,
            'D'     : 0x20,
            'Q'     : 0x10,
            'E'     : 0x12,
            'F'     : 0x21,
            'G'     : 0x22,
            '['     : 0x1A,
            ']'     : 0x1B,
            'SHIFT' : 0x2A,
            'CTRL'  : 0x1D,
            'SPACE' : 0x39}

# C struct redefinitions 
PUL = ctypes.POINTER(ctypes.c_ulong)
class KeyBdInput(ctypes.Structure):
    _fields_ = [("wVk", ctypes.c_ushort),
                ("wScan", ctypes.c_ushort),
                ("dwFlags", ctypes.c_ulong),
                ("time", ctypes.c_ulong),
                ("dwExtraInfo", PUL)]

class HardwareInput(ctypes.Structure):
    _fields_ = [("uMsg", ctypes.c_ulong),
                ("wParamL", ctypes.c_short),
                ("wParamH", ctypes.c_ushort)]

class MouseInput(ctypes.Structure):
    _fields_ = [("dx", ctypes.c_long),
                ("dy", ctypes.c_long),
                ("mouseData", ctypes.c_ulong),
                ("dwFlags", ctypes.c_ulong),
                ("time",ctypes.c_ulong),
                ("dwExtraInfo", PUL)]

class Input_I(ctypes.Union):
    _fields_ = [("ki", KeyBdInput),
                 ("mi", MouseInput),
                 ("hi", HardwareInput)]

class Input(ctypes.Structure):
    _fields_ = [("type", ctypes.c_ulong),
                ("ii", Input_I)]

# Actuals Functions

def PressKey(hexKeyCode):
    extra = ctypes.c_ulong(0)
    ii_ = Input_I()
    ii_.ki = KeyBdInput( 0, hexKeyCode, 0x0008, 0, ctypes.pointer(extra) )
    x = Input( ctypes.c_ulong(1), ii_ )
    ctypes.windll.user32.SendInput(1, ctypes.pointer(x), ctypes.sizeof(x))

def ReleaseKey(hexKeyCode):
    extra = ctypes.c_ulong(0)
    ii_ = Input_I()
    ii_.ki = KeyBdInput( 0, hexKeyCode, 0x0008 | 0x0002, 0, ctypes.pointer(extra) )
    x = Input( ctypes.c_ulong(1), ii_ )
    ctypes.windll.user32.SendInput(1, ctypes.pointer(x), ctypes.sizeof(x))
    

    
#========================================================================================================#

    

pitch = [0, 'P_W', 42, 'R_W', 58, 'P_W', 42, 'R_W']
roll  = [0, 'P_D', 12, 'R_D', 88, 'P_D', 12, 'R_D']
yaw   = [0, 'P_E', 33, 'R_E', 77, 'P_E', 33, 'R_E']
trust = [0, 'P_SHIFT', 5, 'R_SHIFT', 5, 'P_SHIFT', 5, 'R_SHIFT', 5, 'P_SHIFT', 5, 'R_SHIFT', 5, 'P_SHIFT', 5, 'R_SHIFT']
gears = [0, 'P_G', 5, 'R_G']
flaps = [0, 'P_[', 5, 'R_[', 5, 'P_[', 5, 'R_[']

#gears = []

#22

def concatenate(pitch_commands, roll_commands, yaw_commands, trust_commands, gears_commdans, flaps_commands):
    
    res = []
    
    total_commands_ammount = int((len(pitch_commands) + len(roll_commands) + len(yaw_commands) + len(trust_commands) + len(gears_commdans) + len(flaps_commands)) // 2)
    matrix = [pitch_commands, roll_commands, yaw_commands, trust_commands, gears_commdans, flaps_commands]
    
    cursors = [0, 0, 0, 0, 0, 0]
    all_wait_time = 0
    
    for i in range(total_commands_ammount):
        min_wait_time = 100
        min_id = 0
        for j in range(6):
            if cursors[j] <= len(matrix[j]) - 2:
                if matrix[j][cursors[j]] < min_wait_time:
                    min_wait_time = matrix[j][cursors[j]]
                    min_id = j
        res.append(matrix[min_id][cursors[min_id]])
        res.append(matrix[min_id][cursors[min_id]+1])
        all_wait_time += matrix[min_id][cursors[min_id]]
        for j in range(6):
            if j != min_id:
                if cursors[j] <= len(matrix[j]) - 2:
                    matrix[j][cursors[j]] -= matrix[min_id][cursors[min_id]]
        cursors[min_id] += 2
    res.append(200 - all_wait_time)
    
    return res

def use_mechanization(commands):
    for i in commands:
        if type(i) == int and i != 0:
            time.sleep(i/1000)
        
        elif type(i) != int:
            if i[0] == 'P':
                PressKey(KeyCodes[i[2:]])
            elif i[0] == 'R':
                ReleaseKey(KeyCodes[i[2:]])
        
    return 0


                
    
if __name__ == '__main__':
    start_time = time.time()
    res = concatenate(pitch, roll, yaw, trust, gears, flaps)
    use_mechanization(res)
    print(time.time() - start_time)
    print(res)
```
